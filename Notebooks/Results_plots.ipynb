```python
import matplotlib.pyplot as plt
import json
```

```python
results_1 = json.loads(open("result_adam_256.json").read())
results_2 = json.loads(open("result_adam_512.json").read())
results_3 = json.loads(open("result_adam_128.json").read())
results_4 = json.loads(open("result_sgd_128.json").read())

results_5 = json.loads(open("result_adam_256_lr_1.json").read())
results_6 = json.loads(open("result_adam_256_lr_3.json").read())
results_7 = json.loads(open("result_adam_256_lr_6.json").read())

results_8 = json.loads(open("result_adam_512_without_lookahead.json").read())
```

```python
import pandas as pd
df = pd.DataFrame()
df['Adam with 128 BATCH SIZE'] = [results_3[result]["test"]["acc"] for result in results_3.keys()]
df['Adam with 256 BATCH SIZE'] = [results_1[result]["test"]["acc"] for result in results_1.keys()]
df['Adam with 512 BATCH SIZE'] = [results_2[result]["test"]["acc"] for result in results_2.keys()]
```

```python
ax = df.plot(figsize=(20,20),kind='line',marker='x')
ax.set_title("Comparison of Batch Size - Epoch vs Accuracy ") 

ax.set_xlabel("Epochs")
ax.set_ylabel("Test Accuracy")
print("Model : Resnet-18\nLearning Rate : 1e-2\nOptimizer : ADAM\nEpochs : 50")
ax.get_figure().savefig('epoch_vs_accuracy_batch_size.png')
```

```python
import pandas as pd
df2 = pd.DataFrame()
df2['Adam with 128 BATCH SIZE'] = [results_3[result]["test"]["loss"] for result in results_3.keys()]
df2['Adam with 256 BATCH SIZ?E'] = [results_1[result]["test"]["loss"] for result in results_1.keys()]
df2['Adam with 512 BATCH SIZE'] = [results_2[result]["test"]["loss"] for result in results_2.keys()]
```

```python
ax = df2.plot(figsize=(20,20),kind='line',marker='x')
ax.set_title("Comparison of Batch Size - Epoch vs Loss") 
ax.set_xlabel("Epochs")
ax.set_ylabel("Test Loss")
print("Model : Resnet-18\nLearning Rate : 1e-2\nOptimizer : ADAM\nEpochs : 50")
ax.get_figure().savefig('epoch_vs_loss_batch_size.png')
```

```python

df3 = pd.DataFrame()
df3['Adam with Learning Rate 1e-1'] = [results_5[result]["test"]["acc"] for result in results_5.keys()]
df3['Adam with Learning Rate 1e-2'] = [results_1[result]["test"]["acc"] for result in results_1.keys()]
df3['Adam with Learning Rate 1e-3'] = [results_6[result]["test"]["acc"] for result in results_6.keys()]
df3['Adam with Learning Rate 1e-6'] = [results_7[result]["test"]["acc"] for result in results_7.keys()]
```

```python
ax = df3.plot(figsize=(20,20),kind='line',marker='x')
ax.set_title("Comparison of Learning Rate") 
ax.set_xlabel("Epochs")
ax.set_ylabel("Accuracy")
print("Model : Resnet-18\nOptimizer : ADAM \nBatch Size : 256\nEpochs : 50")
ax.get_figure().savefig('Learning_Rate.png')
```

```python
df4 = pd.DataFrame()
df4['ADAM with Lookahead'] = [results_2[result]["test"]["acc"] for result in results_2.keys()]
df4['Adam without Lookahead'] = [results_8[result]["test"]["acc"] for result in results_8.keys()]

ax = df4[20:50].plot(figsize=(20,20),kind='line',marker='x')
ax.set_title("Adam optimizer with vs without Lookahead") 
ax.set_xlabel("Epochs")
ax.set_ylabel("Accuracy")
print("Model : Resnet-18\nOptimizer : Adam with vs without Lookahead \nBatch Size : 256\nEpochs : 50")
ax.get_figure().savefig('adam_with_vs_without_lookahead.png')
```
